```python
"""
Q2.  Explain the difference between mutable and immutable data types with examples?
Ans. Mutable-:
        1. It can be chnaged after creation.
        2. Usually memory address stays the same after change.
        3. For example-> List, Dict.
     Immutable-:
         1. It can not be changed after creation.
         2. A new object is created if modified.
         3. For example-> String, Tuple.
"""
#Example
a=[1,2,3]
print(a)
a[0]=10
print("Mutuable example",a)

#====================================================
b="Hello"
print(b)
b[0]="Y"
print("Immutable Example",b)
```
